```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_json('G_water_1950-01-01_2018-06-22.json')
new_data = data.loc['timeSeries','value']
```

```python
site_array = []
    for elmnt in new_data:
        new_dict = {
            'site_ID':elmnt['sourceInfo']['siteCode'][0]['value'],
            'site_lat': elmnt['sourceInfo']['geoLocation']['geogLocation']['latitude'],
            'site_lng': elmnt['sourceInfo']['geoLocation']['geogLocation']['longitude'],
            'srs': elmnt['sourceInfo']['geoLocation']['geogLocation']['srs'],
            'water_type': elmnt['sourceInfo']['siteProperty'][0]['value'],
            'mmt_type': elmnt['variable']['variableName'],
            'values': elmnt['values']
        }
        
        site_array.append(new_dict)
```

```python
for site in site_array:
    for value in site['values']:
        del value['qualifier']
        del value['qualityControlLevel']
        del value['method']
        del value['source']
        del value['offset']
        del value['sample']
        del value['censorCode']
```

```python
cleaned_site_array = []
for site in site_array:
    for values in site['values']:
        for value in values['value']:
            del value['qualifiers']
            cleaned_site_array.append(
                [site['site_ID'], 
                 site['site_lat'], 
                 site['site_lng'], 
                 site['srs'], 
                 site['water_type'],
                 site['mmt_type'],
                 value['value'],
                 value['dateTime']
            ])
df = pd.DataFrame(cleaned_site_array, columns=['site_ID', 'site_lat', 'site_lng', 'srs', 'water_type', 'mmt_type', 'value', 'dateTime'])
df.head(50)
```

```python
df.to_csv('G_water_1950-01-01_2018-06-22.csv')
```
